# CITRA 1

**SOBEL**

```python
import cv2
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('/content/img1.png')
```

```python
#sobel
img_sobelx = cv2.Sobel(img,cv2.CV_8U,1,0,ksize=5)
img_sobely = cv2.Sobel(img,cv2.CV_8U,0,1,ksize=5)
img_sobel = img_sobelx + img_sobely

# Plotting
plt.imshow(img_sobel)
plt.show()
```

**PREWITT**

```python
#prewitt
kernelx = np.array([[1,1,1],[0,0,0],[-1,-1,-1]])
kernely = np.array([[-1,0,1],[-1,0,1],[-1,0,1]])

img_prewittx = cv2.filter2D(img, -1, kernelx)
img_prewitty = cv2.filter2D(img, -1, kernely)
img_prewitt = img_prewittx + img_prewitty

# Plotting
plt.imshow(img_prewitt)
plt.show()
```

**CANNY**

```python
#canny
img_canny = cv2.Canny(img,100,200)

# Plotting
plt.imshow(img_canny)
plt.show()
```

**LoG**

```python
# Remove noise by blurring with a Gaussian filter
img = cv.GaussianBlur(img, (3, 3), 0)

# Convert the image to grayscale
img_gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

# Apply Laplace function
img_laplace = cv.Laplacian(img_gray, ddepth=cv.CV_16S, ksize=3)

# converting back to uint8
img_log = cv.convertScaleAbs(img_laplace)

# Plotting
plt.imshow(img_log)
plt.show()
```
